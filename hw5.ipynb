### О задании

На занятиях мы подробно обсуждали метод опорных векторов (SVM). В базовой версии в нём нет чего-то особенного — мы всего лишь используем специальную функцию потерь, которая не требует устремлять отступы к бесконечности; ей достаточно, чтобы отступы были не меньше +1. Затем мы узнали, что SVM можно переписать в двойственном виде, который, позволяет заменить скалярные произведения объектов на ядра. Это будет соответствовать построению модели в новом пространстве более высокой размерности, координаты которого представляют собой нелинейные модификации исходных признаков.

Ядровой SVM, к сожалению, довольно затратен по памяти (нужно хранить матрицу Грама размера $d \times d$) и по времени (нужно решать задачу условной оптимизации с квадратичной функцией, а это не очень быстро). Мы обсуждали, что есть способы посчитать новые признаки $\tilde \varphi(x)$ на основе исходных так, что скалярные произведения этих новых $\langle \tilde \varphi(x), \tilde \varphi(z) \rangle$ приближают ядро $K(x, z)$.

Мы будем исследовать аппроксимации методом **Random Fourier Features (RFF, также в литературе встречается название Random Kitchen Sinks)** для гауссовых ядер. Будем использовать формулы, которые немного отличаются от того, что было на лекциях (мы добавим сдвиги внутрь тригонометрических функций и будем использовать только косинусы, потому что с нужным сдвигом косинус превратится в синус):
$$\tilde \varphi(x) = (
\cos (w_1^T x + b_1),
\dots,
\cos (w_n^T x + b_n)
),$$
где $w_j \sim \mathcal{N}(0, 1/\sigma^2)$, $b_j \sim U[-\pi, \pi]$, где

$1/\sigma^2$ --- **дисперсия распределения.**

На новых признаках $\tilde \varphi(x)$ мы будем строить любую линейную модель.

Можно считать, что это некоторая новая парадигма построения сложных моделей. Можно направленно искать сложные нелинейные закономерности в данных с помощью градиентного бустинга или нейронных сетей, а можно просто нагенерировать большое количество случайных нелинейных признаков и надеяться, что быстрая и простая модель (то есть линейная) сможет показать на них хорошее качество. В этом задании мы изучим, насколько работоспособна такая идея.

### Алгоритм

Вам потребуется реализовать следующий алгоритм:
1. Понизить размерность выборки до new_dim с помощью метода главных компонент.

2. Для полученной выборки оценить гиперпараметр $\sigma^2$ с помощью эвристики (рекомендуем считать медиану не по всем парам объектов, а по случайному подмножеству из где-то миллиона пар объектов): $$\sigma^2 = \text{median}_{i, j = 1, \dots, \ell, i \neq j} \left\{\sum_{k = 1}^{d} (x_{ik} - x_{jk})^2 \right\}$$
**Замечание:** обратите внимание на  $i \neq j$, без этого оценка медианы может быть смещена, а также без этого будут сниматься баллы.

3. Сгенерировать n_features наборов весов $w_j$ и сдвигов $b_j$.

4. Сформировать n_features новых признаков по формулам, приведённым выше.

5. Обучить линейную модель (логистическую регрессию или SVM) на новых признаках.

6. Повторить преобразования (PCA, формирование новых признаков) к тестовой выборке и применить модель.

**Подсказка**
* Использование метода без PCA и без нормализации признаков перед генерацией весов может дать неожиданные результаты. Сами подумайте, почему так может происходить :)

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.decomposition import PCA
from scipy.spatial import distance_matrix
from sklearn.preprocessing import FunctionTransformer, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
import time


class RFFPipeline(BaseEstimator, TransformerMixin):
    def __init__(self, n_features=1000, new_dim=50, use_PCA=True, classifier='logreg', random_state=0, **kwargs):
        """        
        Implements pipeline, which consists of PCA decomposition,
        Random Fourier Features approximation and linear classification model.
        
        n_features, int: amount of synthetic random features generated with RFF approximation.

        new_dim, int: PCA output size.
        
        use_PCA, bool: whether to include PCA preprocessing.
        
        classifier, string: either 'svm' or 'logreg', a linear classification model to use on top of pipeline.
        
        Feel free to edit this template for your preferences.    
        """
        self.random_state = random_state
        self.n_features = n_features
        self.use_PCA = use_PCA
        self.new_dim = new_dim
        self.classifier = classifier
        self.kwargs = kwargs
        

    def fit(self, X, y):
        """
        Fit all parts of algorithm (PCA, RFF, Classification) to training set.
        """
        np.random.seed(self.random_state)
        # Понизить размерность выборки до new_dim с помощью метода главных компонент.
        start = time.time()
        self.scaler = StandardScaler()
        X_new = self.scaler.fit_transform(X)
        if self.use_PCA:
            self.pca = PCA(n_components=self.new_dim)
            X_new = self.pca.fit_transform(X_new)
        self.preprocess_time = time.time() - start
        start = time.time()
        # Для полученной выборки оценить гиперпараметр σ^2 с помощью эвристики
        index = np.random.choice(X_new.shape[0], min(X_new.shape[0], 1000), replace=False)
        sigma = np.median(distance_matrix(X_new[index], X_new[index]))
        
        # Сгенерировать n_features наборов весов w_j и сдвигов b_j
        # w_j ~ N(0, 1/sigma), b_j ~ U[-pi, pi]
        w = np.random.normal(0, 1 / sigma, (X_new.shape[1], self.n_features))
        b = np.random.uniform(-np.pi, np.pi, self.n_features)
        
        # Сформировать n_features новых признаков по формулам, приведённым выше.
        RFF_transformer = FunctionTransformer(lambda Z: np.cos(np.dot(Z, w) + b))
        X_new = RFF_transformer.transform(X_new)
        
        # Обучить линейную модель (логистическую регрессию или SVM) на новых признаках.
        if self.classifier == 'logreg':
            self.model = LogisticRegression(**self.kwargs)
        elif self.classifier == 'svm':
            self.model = LinearSVC(**self.kwargs)
        else:
            raise NotImplementedError
        
        self.model.fit(X_new, y)
        self.fit_time = time.time() - start
        self.RFF_transformer = RFF_transformer
        return self

    def predict_proba(self, X):
        """
        Apply pipeline to obtain scores for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        if self.classifier == 'logreg':
            return self.model.predict_proba(X)
        elif self.classifier == 'svm':
            return self.model.decision_function(X)

    def predict(self, X):
        """
        Apply pipeline to obtain discrete predictions for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        return self.model.predict(X)
```

Многие алгоритмы машинного обучения работают лучше, если признаки данных некоррелированы. Оказывается, что для RFF существует модификация, позволяющая получать ортогональные случайные признаки (Orthogonal Random Features, ORF). Об этом методе можно прочитать в [статье](https://arxiv.org/abs/1610.09072#:~:text=We%20call%20this%20technique%20Orthogonal,to%20speed%20up%20the%20computation). Реализуйте класс для вычисления ORF по аналогии с основным заданием. Обратите внимание, что ваш класс должен уметь работать со случаем n_features > new_dim (в статье есть замечание на этот счет). Проведите эксперименты, сравнивающие RFF и ORF, **(?)** сделайте выводы.

**Замечание:**
* Сравнения делайте как по времени, так и по качеству.
* Можно сравнивать качество двух методов на разном числе признаков, в таком случае очень полезным будет график зависимости качества от числа признаков

```python
from scipy.linalg import qr
from scipy.spatial.distance import pdist

class ORFPipeline(BaseEstimator, TransformerMixin):
    def __init__(self, n_features=1000, new_dim=50, use_PCA=True, classifier='logreg', random_state=0, **kwargs):
        """        
        Implements pipeline, which consists of PCA decomposition,
        Random Fourier Features approximation and linear classification model.
        
        n_features, int: amount of synthetic random features generated with RFF approximation.

        new_dim, int: PCA output size.
        
        use_PCA, bool: whether to include PCA preprocessing.
        
        classifier, string: either 'svm' or 'logreg', a linear classification model to use on top of pipeline.
        
        Feel free to edit this template for your preferences.    
        """
        self.random_state = random_state
        self.n_features = n_features
        self.use_PCA = use_PCA
        self.new_dim = new_dim
        self.classifier = classifier
        self.kwargs = kwargs
        

    def fit(self, X, y):
        """
        Fit all parts of algorithm (PCA, RFF, Classification) to training set.
        """
        np.random.seed(self.random_state)
        # Понизить размерность выборки до new_dim с помощью метода главных компонент.
        start = time.time()
        self.scaler = StandardScaler()
        X_new = self.scaler.fit_transform(X)
        if self.use_PCA:
            self.pca = PCA(n_components=self.new_dim)
            X_new = self.pca.fit_transform(X_new)
        self.preprocess_time = time.time() - start
        start = time.time()
        # Для полученной выборки оценить гиперпараметр σ^2 с помощью эвристики
        index = np.random.choice(X_new.shape[0], min(X_new.shape[0], 1000), replace=False)
        sigma = np.median(distance_matrix(X_new[index], X_new[index]))
        # Сгенерировать n_features наборов весов w_j и сдвигов b_j
        b = np.random.uniform(-np.pi, np.pi, self.n_features)
        
        w = None
        d = X_new.shape[1]
        for i in range(0, self.n_features, d):
            G = np.random.normal(0, 1, (d, d))
            Q, R = qr(G) # where Q is unitary/orthogonal and R upper triangular
            S = np.diagflat(np.random.chisquare(d, d))
            new_weights = np.dot(np.sqrt(S), Q) / (sigma)
            if i == 0:
                w = new_weights
            else:
                w = np.concatenate((w, new_weights), axis=1)

        w = w[:, :self.n_features]
        
        # Сформировать n_features новых признаков по формулам, приведённым выше.
        RFF_transformer = FunctionTransformer(lambda Z: np.cos(np.dot(Z, w) + b))
        X_new = RFF_transformer.transform(X_new)
        
        # Обучить линейную модель (логистическую регрессию или SVM) на новых признаках.
        if self.classifier == 'logreg':
            self.model = LogisticRegression(**self.kwargs)
        elif self.classifier == 'svm':
            self.model = LinearSVC(**self.kwargs)
        else:
            raise NotImplementedError
        
        self.model.fit(X_new, y)
        self.fit_time = time.time() - start
        self.RFF_transformer = RFF_transformer
        return self

    def predict_proba(self, X):
        """
        Apply pipeline to obtain scores for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        if self.classifier == 'logreg':
            return self.model.predict_proba(X)
        elif self.classifier == 'svm':
            return self.model.decision_function(X)

    def predict(self, X):
        """
        Apply pipeline to obtain discrete predictions for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        return self.model.predict(X)
```

Рассмотрим ORF для логистической регрессии

```python
class ORFPipeline_func(BaseEstimator, TransformerMixin):
    def __init__(self, n_features=1000, new_dim=50, use_PCA=True, classifier='logreg', 
                 random_state=0, func=np.cos, **kwargs):
        """        
        Implements pipeline, which consists of PCA decomposition,
        Random Fourier Features approximation and linear classification model.
        
        n_features, int: amount of synthetic random features generated with RFF approximation.

        new_dim, int: PCA output size.
        
        use_PCA, bool: whether to include PCA preprocessing.
        
        classifier, string: either 'svm' or 'logreg', a linear classification model to use on top of pipeline.
        
        Feel free to edit this template for your preferences.    
        """
        self.random_state = random_state
        self.func = func
        self.n_features = n_features
        self.use_PCA = use_PCA
        self.new_dim = new_dim
        self.classifier = classifier
        self.kwargs = kwargs
        

    def fit(self, X, y):
        """
        Fit all parts of algorithm (PCA, RFF, Classification) to training set.
        """
        np.random.seed(self.random_state)
        # Понизить размерность выборки до new_dim с помощью метода главных компонент.
        start = time.time()
        self.scaler = StandardScaler()
        X_new = self.scaler.fit_transform(X)
        if self.use_PCA:
            self.pca = PCA(n_components=self.new_dim)
            X_new = self.pca.fit_transform(X_new)
        self.preprocess_time = time.time() - start
        start = time.time()
        # Для полученной выборки оценить гиперпараметр σ^2 с помощью эвристики
        index = np.random.choice(X_new.shape[0], min(X_new.shape[0], 1000), replace=False)
        sigma = np.median(distance_matrix(X_new[index], X_new[index]))
        # Сгенерировать n_features наборов весов w_j и сдвигов b_j
        b = np.random.uniform(-np.pi, np.pi, self.n_features)
        
        w = None
        d = X_new.shape[1]
        for i in range(0, self.n_features, d):
            G = np.random.normal(0, 1, (d, d))
            Q, R = qr(G) # where Q is unitary/orthogonal and R upper triangular
            S = np.diagflat(np.random.chisquare(d, d))
            new_weights = np.dot(np.sqrt(S), Q) / (sigma)
            if i == 0:
                w = new_weights
            else:
                w = np.concatenate((w, new_weights), axis=1)

        w = w[:, :self.n_features]
        
        # Сформировать n_features новых признаков по формулам, приведённым выше.
        RFF_transformer = FunctionTransformer(lambda Z: self.func(np.dot(Z, w) + b))
        
        X_new = RFF_transformer.transform(X_new)
        
        # Обучить линейную модель (логистическую регрессию или SVM) на новых признаках.
        if self.classifier == 'logreg':
            self.model = LogisticRegression(**self.kwargs)
        elif self.classifier == 'svm':
            self.model = LinearSVC(**self.kwargs)
        elif self.classifier == 'LGBM':
            self.model = LGBMClassifier(learning_rate=0.1, n_estimators=350, **self.kwargs)
        else:
            raise NotImplementedError
        
        self.model.fit(X_new, y)
        self.fit_time = time.time() - start
        self.RFF_transformer = RFF_transformer
        return self

    def predict_proba(self, X):
        """
        Apply pipeline to obtain scores for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        if (self.classifier == 'logreg') or (self.classifier == 'LGBM'):
            return self.model.predict_proba(X)
        elif self.classifier == 'svm':
            return self.model.decision_function(X)

    def predict(self, X):
        """
        Apply pipeline to obtain discrete predictions for input data.
        """
        # Your code here: (￣▽￣)/♫•*¨*•.¸¸♪
        X = self.scaler.transform(X)
        if self.use_PCA:
            X = self.pca.transform(X)
        X = self.RFF_transformer.transform(X)
        return self.model.predict(X)
```
